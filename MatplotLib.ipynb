```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data = pd.read_csv('RS_Session_250_AU641.csv')
```

```python
display(data)
```

```python
data2 = data.head(len(data)-1)
data2 = data2.dropna(subset=['Total Foodgrains - 2016-17', 'Total Foodgrains - 2017-18'], thresh=1)
display(data2)
```

```python
from sklearn.linear_model import LinearRegression
# data2 = data.head(len(data)-1)
fig, ax = plt.subplots()
y1 = data2['Total Foodgrains - 2016-17'].values
y2 = data2['Total Foodgrains - 2017-18'].values
ax.scatter(y1, y2)
# ax.plot([y1.min(), y1.max()], [y1.min(), y1.max()], 'k--', lw=4)
ax.set_xlabel('Total Foodgrains - 2016-17')
ax.set_ylabel('Total Foodgrains - 20170')
#regression line
y1, y2 = y1.reshape(-1,1), y2.reshape(-1,1)
ax.plot(y1, LinearRegression().fit(y1, y2).predict(y2), '--')
plt.legend(data2['States/UT'])
plt.show()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
colormap = cm.plasma
colorlist = [colors.rgb2hex(colormap(i)) for i in np.linspace(0, 0.9, len(data2))]
# plt.figure(figsize=(40,40))
fig, ax = plt.subplots(figsize = (20,20))
for i,c in enumerate(colorlist):

    x = data2['Total Foodgrains - 2016-17'][i]
    y = data2['Total Foodgrains - 2017-18'][i]
    l = data2['States/UT'][i]

    ax.scatter(x, y, label=l, s=50, linewidth=10, c=c)
ax.legend(fontsize='xx-large', ncol=3)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.tick_params(axis='both', which='minor', labelsize=13)
plt.xlabel('Total Foodgrains produced in kilotons - StateWise - 2016-17', fontsize=25)
plt.ylabel('Total Foodgrains produced in kilotons - Statewise - 2017-18', fontsize=25)
plt.show()
```

```python
cols = [x for x in data.columns if 'Total' in x]
print(cols)
fig, ax = plt.subplots(figsize = (20,20))
ax.boxplot(data2[cols[:2]], boxprops=dict(color='red'),labels=cols[:2])
ax.tick_params(axis='both', which='major', labelsize=25)
ax.tick_params(axis='both', which='minor', labelsize=13)
plt.ylabel('Total Food grains produce in kilotones', fontsize=25)
```
